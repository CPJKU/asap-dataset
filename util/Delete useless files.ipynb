```python
import pandas as pd
from pathlib import Path
import numpy as np

BASE_PATH = "../"
```

```python
def perf_annotation_path(row):
    if Path(BASE_PATH, row["midi_performance"][:-4]+"_ann_cleaned.txt").exists():
        return row["midi_performance"][:-4]+"_ann_cleaned.txt"
    elif Path(BASE_PATH, row["midi_performance"][:-4]+"_ann.txt").exists():
        return row["midi_performance"][:-4]+"_ann.txt"
    else:
        print("No annotations for",row.midi_performance)
```

```python
all_df = pd.read_csv(Path(BASE_PATH,"ASAP_v2.csv"))
all_df["performance_annotations"] = all_df.apply(perf_annotation_path,axis=1)
all_df
```

# About the scores

```python
quant_df = all_df.drop_duplicates(subset="midi_score_annotations",keep="first")
```

```python
# change the xml score path
for i,row in quant_df.iterrows():
    xml_score = Path(BASE_PATH,row["xml_score"])
    xml_score.rename(Path(xml_score.parent,"xml_score.musicxml"))
```

```python
# change the midi score path
for i,row in quant_df.iterrows():
    midi_score = Path(BASE_PATH,row["midi_score"])
    midi_score.rename(Path(midi_score.parent,"midi_score.mid"))
```

```python
# change the annotations score path
for i,row in quant_df.iterrows():
    midi_score_annotations = Path(BASE_PATH,row["midi_score_annotations"])
    midi_score_annotations.rename(Path(midi_score_annotations.parent,"midi_score_annotations.txt"))
```

```python
# update the positions in the table
def update_xml_score_name(row):
    xml_score = row["xml_score"]
    new_path = xml_score[:-25] + "xml_score.musicxml"
    assert(Path(BASE_PATH,new_path).exists())
    return new_path

all_df["xml_score"] = all_df.apply(update_xml_score_name,axis=1)
```

```python
def update_midi_score_name(row):
    midi_score = row["midi_score"]
    new_path = midi_score[:-16] + "midi_score.mid"
    assert(Path(BASE_PATH, new_path).exists())
    return new_path

all_df["midi_score"] = all_df.apply(update_midi_score_name,axis=1)
```

```python
def update_midi_score_annotations_name(row):
    midi_score_annotations = row["midi_score_annotations"]
    new_path = midi_score_annotations[:-15] + "midi_score_annotations.txt"
    assert(Path(BASE_PATH,new_path).exists())
    return new_path

all_df["midi_score_annotations"] = all_df.apply(update_midi_score_annotations_name,axis=1)
```

# About the performances

```python
# change the annotations score path
for i,row in all_df.iterrows():
    midi_performance = Path(BASE_PATH,row["midi_performance"])
    performance_annotations = Path(BASE_PATH,row["performance_annotations"])
    performance_annotations.rename(Path(midi_performance.parent,midi_performance.stem + "_annotations.txt"))
```

```python
def update_midi_performance_annotations_name(row):
    midi_performance = row["midi_performance"]
    midi_performance_annotations = Path(BASE_PATH,row["performance_annotations"])
    new_path = midi_performance[:-4] + "_annotations.txt"
    assert(Path(BASE_PATH,new_path).exists())
    return new_path

all_df["performance_annotations"] = all_df.apply(update_midi_performance_annotations_name,axis=1)
```

```python
all_df.to_csv('ASAP_v3.csv', index = False)
```

# Chek if the performance is cleaned

```python
def is_manually_checked(row):
    if Path(BASE_PATH,row["midi_performance"][:-4]+ "_ann.txt").exists():
        return True
    else:
        return False
    
```

```python
all_df["is_manually_checked"] = all_df.apply(is_manually_checked,axis = 1)
```

```python
# remove the click for the manually checked annotations
for i, row in all_df[all_df["is_manually_checked"]].iterrows():
    Path(BASE_PATH,row["midi_performance"][:-4]+ "_click.wav").unlink()
```
